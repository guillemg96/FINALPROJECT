INTRODUCCIÓN                                  
En este proyecto, se analizarán los municipios de Valladolid, España, con el objetivo de determinar la viabilidad de abrir un nuevo restaurante. Valladolid es conocida por su destacada oferta gastronómica, y utilizando datos y la API de Foursquare, se examinará el crecimiento de los municipios en la periferia y la competencia existente. Mediante técnicas de Machine Learning, se pronosticará el potencial de crecimiento de los municipios en los próximos años para respaldar la toma de decisiones. Se presentará un resultado objetivo y se justificará por qué abrir un restaurante en la provincia sería una opción óptima, lo cual puede resultar útil para inversores y propietarios de negocios gastronómicos en busca de una ubicación.


DATOS
Se utilizarán datos obtenido de webs oficiales para darle veracidad al proyecto, como la web de ayuntamiento de Valladolid y datos.gob.es
También se utilizará Foursquare para localizar por coordenadas.
